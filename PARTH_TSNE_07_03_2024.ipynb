# DATA CLEANING

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

# Load the CSV file into a DataFrame
df = pd.read_csv('ticket_dump.csv')

# Function to clean and preprocess text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Tokenization
    tokens = word_tokenize(text)

#     # Remove stop words
#     stop_words = set(stopwords.words('english'))
#     tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(tokens)

# Apply text preprocessing to the 'Description Text' column
df['Cleaned_Description_Text'] = df['Description Text'].apply(preprocess_text)

# Save the updated DataFrame to a new CSV file
df.to_csv('cleaned_ticket_dump.csv', index=False)

# Display the first few rows of the updated DataFrame
print(df[['Description Text', 'Cleaned_Description_Text']].head())
```

# APPLYING BERT EMBEDDINGS

```python
import pandas as pd
from transformers import BertModel, BertTokenizer
import torch
from tqdm import tqdm

# Load the cleaned CSV file into a DataFrame
df = pd.read_csv('cleaned_ticket_dump.csv')

# Load pre-trained BERT model and tokenizer
model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Function to obtain BERT embeddings
def get_bert_embeddings(text):
    # Tokenize the text
    tokens = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)

    # Forward pass through the BERT model
    with torch.no_grad():
        outputs = model(**tokens)

    # Extract the embeddings for the [CLS] token (first token)
    embeddings = outputs.last_hidden_state[:, 0, :].squeeze().numpy()

    return embeddings

# Apply BERT embeddings to the 'Cleaned_Description_Text' column with a progress bar
tqdm.pandas(desc="Applying BERT Embeddings")
df['BERT_Embeddings'] = df['Cleaned_Description_Text'].progress_apply(get_bert_embeddings)

# Save the DataFrame with BERT embeddings to a Parquet file
df.to_parquet('bert_embeddings_ticket_dump.parquet', index=False)

# Display the first few rows of the updated DataFrame
print(df[['Cleaned_Description_Text', 'BERT_Embeddings']].head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from tqdm import tqdm

# Load the DataFrame from the Parquet file
df = pd.read_parquet('bert_embeddings_ticket_dump.parquet')

# Extract BERT embeddings as a numpy array
embeddings_array = np.array(df['BERT_Embeddings'].tolist())

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
embeddings_tsne = tsne.fit_transform(embeddings_array)

# Visualize the results
plt.scatter(embeddings_tsne[:, 0], embeddings_tsne[:, 1])
plt.title("t-SNE Visualization of BERT Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.savefig('tsne_plot.png')  # Save the t-SNE plot

# Assuming you want to cluster similar tickets
# Aggregating embeddings based on TicketID (mean aggregation)
agg_embeddings = df.groupby('TicketID')['BERT_Embeddings'].agg('mean').tolist()
agg_embeddings_array = np.array(agg_embeddings)

# Apply KMeans for clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Adjust the number of clusters as needed
cluster_labels = kmeans.fit_predict(agg_embeddings_array)

# Add cluster labels to your DataFrame
cluster_df = pd.DataFrame({'TicketID': df['TicketID'].unique(), 'Cluster_Labels': cluster_labels})
df = pd.merge(df, cluster_df, on='TicketID')

# Use TF-IDF to extract representative terms for each cluster
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Cleaned_Description_Text'])

terms = tfidf_vectorizer.get_feature_names_out()

cluster_representative_terms = {}
for cluster_id in tqdm(range(max(cluster_labels) + 1), desc="Extracting Representative Terms"):
    cluster_indices = df[df['Cluster_Labels'] == cluster_id].index
    cluster_tfidf = tfidf_matrix[cluster_indices].sum(axis=0).A1
    top_terms_indices = cluster_tfidf.argsort()[-5:][::-1]  # Adjust the number of top terms as needed
    cluster_representative_terms[cluster_id] = [terms[i] for i in top_terms_indices]

    # Debugging prints
    print(f"\nCluster {cluster_id} - Top Terms: {', '.join([terms[i] for i in top_terms_indices])}")
    print(f"Cluster {cluster_id} - Term Frequencies: {cluster_tfidf[top_terms_indices]}")

# Save cluster information to a file
with open('cluster_info.txt', 'w') as file:
    for cluster_id, terms in cluster_representative_terms.items():
        file.write(f"\nCluster {cluster_id} - Representative Terms: {', '.join(terms)}\n")

# Save the DataFrame with cluster information to a Parquet file
df.to_parquet('clustered_tickets_info.parquet', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from tqdm import tqdm
from transformers import BertTokenizer, BertModel
import torch

# Load the DataFrame from the Parquet file
df = pd.read_parquet('bert_embeddings_ticket_dump.parquet')

# Extract BERT embeddings as a numpy array
bert_embeddings_array = np.array(df['BERT_Embeddings'].tolist())

# Apply t-SNE for dimensionality reduction on BERT embeddings
tsne = TSNE(n_components=2, random_state=42)
bert_embeddings_tsne = tsne.fit_transform(bert_embeddings_array)

# Visualize the results
plt.scatter(bert_embeddings_tsne[:, 0], bert_embeddings_tsne[:, 1])
plt.title("t-SNE Visualization of BERT Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.savefig('tsne_plot.png')  # Save the t-SNE plot

# Assuming you want to cluster similar tickets based on BERT embeddings
kmeans = KMeans(n_clusters=5, random_state=42)  # Adjust the number of clusters as needed
bert_cluster_labels = kmeans.fit_predict(bert_embeddings_array)

# Add BERT cluster labels to your DataFrame
df['BERT_Cluster_Labels'] = bert_cluster_labels

# Use TF-IDF to extract representative terms for each BERT cluster
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Cleaned_Description_Text'])

terms = tfidf_vectorizer.get_feature_names_out()

bert_cluster_representative_terms = {}
for cluster_id in tqdm(range(max(bert_cluster_labels) + 1), desc="Extracting BERT Representative Terms"):
    cluster_indices = df[df['BERT_Cluster_Labels'] == cluster_id].index
    cluster_tfidf = tfidf_matrix[cluster_indices].sum(axis=0).A1
    top_terms_indices = cluster_tfidf.argsort()[-5:][::-1]  # Adjust the number of top terms as needed
    bert_cluster_representative_terms[cluster_id] = [terms[i] for i in top_terms_indices]

    # Debugging prints
    print(f"\nBERT Cluster {cluster_id} - Top Terms: {', '.join([terms[i] for i in top_terms_indices])}")
    print(f"BERT Cluster {cluster_id} - Term Frequencies: {cluster_tfidf[top_terms_indices]}")

# Save BERT cluster information to a file
with open('bert_cluster_info.txt', 'w') as file:
    for cluster_id, terms in bert_cluster_representative_terms.items():
        file.write(f"\nBERT Cluster {cluster_id} - Representative Terms: {', '.join(terms)}\n")

# Save the DataFrame with BERT cluster information to a Parquet file
df.to_parquet('bert_clustered_tickets_info.parquet', index=False)
```

```python
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd
from bertopic import BERTopic
from tqdm import tqdm

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Tokenize and get contextual embeddings for each word
def get_contextual_embeddings(text):
    tokens = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = bert_model(**tokens)
    embeddings = outputs.last_hidden_state.mean(dim=1)  # Use mean pooling for sentence embedding
    return embeddings.numpy()

# Load a subset of your data and preprocess it
subset_size = 100  # Set the desired subset size
df_subset = pd.read_parquet('bert_embeddings_ticket_dump.parquet').head(subset_size)

# Apply contextual embeddings to each document with tqdm progress bar
tqdm.pandas(desc="Applying Contextual Embeddings")
df_subset['contextual_embeddings'] = df_subset['Cleaned_Description_Text'].progress_apply(get_contextual_embeddings)

# Convert embeddings to string
df_subset['contextual_embeddings'] = df_subset['contextual_embeddings'].apply(lambda x: ' '.join(map(str, x.flatten())))

# Initialize BERTopic model with adjusted hyperparameters
topic_model = BERTopic(
    language="english",
    calculate_probabilities=True,
    verbose=True,
    nr_topics=10,  # Adjust this based on your data
    low_memory=True  # Add low_memory parameter to handle potential memory issues
)

# Fit BERTopic model using the contextual embeddings
topics, _ = topic_model.fit_transform(df_subset['contextual_embeddings'].tolist())

# Save BERTopic model to Parquet file
topic_model.save("bertopic_model_subset")

# Save the DataFrame with BERTopic results to a Parquet file
df_subset.to_parquet('bertopic_results_subset.parquet', index=False)
```

```python
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd
from bertopic import BERTopic
from tqdm.notebook import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from hdbscan import HDBSCAN

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Function to get contextual embeddings for each text
def get_contextual_embeddings(text):
    tokens = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = bert_model(**tokens)
    embeddings = outputs.last_hidden_state.mean(dim=1)  # Use mean pooling for sentence embedding
    return embeddings.numpy().flatten()

# Load and preprocess the full data
df = pd.read_parquet('bert_embeddings_ticket_dump.parquet')

# Apply contextual embeddings to each document with tqdm progress bar
tqdm.pandas(desc="Applying Contextual Embeddings")
df['contextual_embeddings'] = df['Cleaned_Description_Text'].progress_apply(get_contextual_embeddings)

# Apply TF-IDF to the full data
tfidf_vectorizer = TfidfVectorizer()
dense_tfidf_matrix = tfidf_vectorizer.fit_transform(df['Cleaned_Description_Text']).toarray()

# Ensure the size of dense_tfidf_matrix matches the size of the BERT embeddings array
if dense_tfidf_matrix.shape[0] != df.shape[0]:
    raise ValueError("The size of dense_tfidf_matrix does not match the size of BERT embeddings array.")

# Combine BERT embeddings and TF-IDF for the full data
combined_embeddings = np.concatenate([df['contextual_embeddings'].to_list(), dense_tfidf_matrix], axis=1)

# Hyperparameter Tuning Section for the full data
# Use hyperparameters from the first code
bertopic_params = {
    "embedding_model": bert_model,
    "vectorizer_model": tfidf_vectorizer,
    "hdbscan_model": HDBSCAN(min_cluster_size=15, metric='euclidean', min_samples=1, cluster_selection_method='leaf')
}

# Initialize BERTopic model with adjusted hyperparameters for the full data
topic_model = BERTopic(**bertopic_params)

# Fit BERTopic model using the text data in the full dataset
topics, _ = topic_model.fit_transform(df['Cleaned_Description_Text'].tolist())

# Get information about the generated topics for the full data
topic_info = topic_model.get_topic_info()

# Map topic names to the original topics
topic_names = {
    -1: "General Information",
    0: "Assistance and Helpdesk",
    1: "Approved Amounts and Remarks",
    2: "Approval Proof and Status",
    3: "Approved Tuition Fees",
    4: "Approved Amounts and Sections",
    5: "Rent Approval and Agreements",
    6: "Feel Free to Reach Assistance",
    7: "Rent, Revenue, and Landlord Agreement",
    8: "FY24, Country, TC, LAC",
    9: "VPF, PF, IE, Reflection",
    10: "Deducted Salary and Contribution",
    11: "Microsoft Message Encryption",
    12: "Ankit Bangerwa and Tata Communication",
    13: "Loan, Home, Property, Interest",
    14: "Password, Offboarding, and Reset",
    15: "Event, Ali Bakhsh, and Team Update",
    16: "Regime, New Tax, and 115bac",
    17: "Approval Proof and Investment Status",
    18: "IPSF, FY24, and Few Reminders",
    19: "Approved Premium and Approver",
    20: "Credited Salary - Tata Bagrecha",
    21: "Fee, Tuition, and Development",
    22: "Approved Amount and Housing Interest",
    23: "Investment Proof and Pandey",
    24: "Medical 80 of 80ddb"
}

df['Topic_Name'] = topics  # Directly create a new column for topics

# Add topic information to the DataFrame for the full data
df[['Count', 'Name', 'Representation', 'Representative_Docs']] = topic_info[['Count', 'Name', 'Representation', 'Representative_Docs']]

# Save the DataFrame with topics and topic names for the full data
df.to_parquet('bert_embeddings_with_topics.parquet', index=False)
```

```python
import plotly.express as px
from sklearn.manifold import TSNE

def visualize_tsne(embeddings, topic_column, topic_names, title='t-SNE Visualization of Topics'):
    # Apply t-SNE to reduce dimensionality
    tsne_model = TSNE(n_components=2, random_state=42)
    tsne_embeddings = tsne_model.fit_transform(embeddings)

    # Create a DataFrame for visualization
    df_tsne = pd.DataFrame(tsne_embeddings, columns=['tsne_1', 'tsne_2'])
    df_tsne[topic_column] = df[topic_column]  # Assuming 'topic' is the column in your DataFrame with topic assignments

    # Map topic assignments to topic names
    df_tsne['topic_name'] = df_tsne[topic_column].map(topic_names)

    # Visualize the data using Plotly Express
    fig = px.scatter(df_tsne, x='tsne_1', y='tsne_2', color='topic_name', title=title)
    fig.show()

# Assuming 'combined_embeddings' is your combined embeddings array
# (the variable you created by concatenating BERT embeddings and TF-IDF features)
visualize_tsne(combined_embeddings, 'Topic_Name', topic_names, title='t-SNE Visualization of Topics')
```
