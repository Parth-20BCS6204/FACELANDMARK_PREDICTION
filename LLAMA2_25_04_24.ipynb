```python
import pandas as pd
import re

def clean_description(text):
    text = text.lower()  # Convert everything to lowercase
    substrings_to_remove = [
        'This is a system generated mail.',
        ' do not reply.',
        ' pvt ltd',
        'Prism ID',
        'DPA',
        'Prabhadevi Industrial Estate',
        'Opp Siddhivinayak Mandir',
        ' Veer Savarkar Marg',
        'prabhadevi mumbai',
        '_',
        ';',
        'tcoc   dwp   posh   isac   tcoc   dwp ',
        'customer service executive phone extn. toll free no head office / palm spring centre new link rd above croma malad w mumbai . branches thane pune hyderabad chennai delhi kolkata bangalore website www.promptpersonnel .com',
        'if you are unable to see the pictures in the mail then right click on the picture and click on download pictures',
        'agency contractor jobrequisition for agency contractor network operations  support is in your prism inbox awaiting your',
        'agency contractor jobrequisition for consultant business collaboration is in your prism inbox awaiting your ',
        'wordbreak breakall  be advised that the document agency contractor jobrequisition for consultant business collaboration is now available in your prism inbox folder for your action.  you can access this document at the following url document link',
        'welcome@',
        '@',
        '#',
        '"',
        '"',
        'this is an autogenerated response. do not reply to this email.',
        'if you are unable to see pictures in the mail right click on the picture and click on download pictures',
        'â',
        '€',
        '™s',
        '™',
        'img src= alt=prism title=prism width= height= border= align=right /',
        'this is an autogenerated message sent by the prism application. do not reply to this email',
        '',
        '',
        '',
        '',
        '',
        '',
        '',
        '',
        '',
        '',
        '',
    ]


    # Remove specific phrases
    text = re.sub(r'dear team\s*,*', '', text, flags=re.IGNORECASE)
    text = re.sub(r'\b\d+\b', '', text)
    text = re.sub(r'thank you\s*,*', '', text, flags=re.IGNORECASE)
    text = re.sub(r'\b(?:cid)?[a-f\d]+\b', '', text)
    text = re.sub(r'\b\d+\b', '', text)
    text = re.sub(r'Dear\s*,*', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'\d{3,}\s*,\s*[\w\s]+,\s*[\w\s]+,\s*[\w\s]+\s*,\s*\w+\s*\d{3}\s*\d{3}\s*\.?', '', text) # will remove addresses from the data
    text = re.sub(r'hello\s*,*', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'greetings\s*,*', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'please\s*,*', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'thanks\s*,*', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'\bhi\b\s*,*', '', text, flags=re.IGNORECASE)
    text = re.sub(r'regards\s*,*', '', text, flags=re.IGNORECASE)
    text = re.sub(r'thanking you\s*,*', '', text, flags=re.IGNORECASE)
    # text = re.sub(r'actionanothref\^=\'mailto\'', text, flags=re.IGNORECASE)

    # Remove dates
    text = re.sub(r'\b\d{1,2}(?:st|nd|rd|th)?\s+(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s+\d{2,4}\b', '', text)

    # Remove punctuation errors and general errors
    text = re.sub(r'(\.{2,}|-+|,+)', '', text)
    text = re.sub(r'(\.{2,}|-+|,+)', ' ', text)
    text = re.sub(r'\b\w+@\w+\.\w+\b', '', text)  # Remove emails
    text = re.sub(r'https?://\S+', '', text)  # Remove website links
    text = re.sub(r'[\[\]{}()]', '', text)  # Remove brackets
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces and newlines
    text = re.sub(r'\b\d+\b', '', text)  # Remove numbers

    # Remove unnecessary phrases
    text = re.sub(r'Dear team', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'|', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'>', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'%', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'<', '', str(text), flags=re.IGNORECASE)
    text = re.sub(r'Dear Team', '', text, flags=re.IGNORECASE)

    # Remove salutations
    text = re.sub(r'Dear.*?ji', '', text, flags=re.IGNORECASE)

    # Remove specific phrases and patterns
    text = re.sub(r'My cources were completed, please check it', '', text, flags=re.IGNORECASE)
    text = re.sub(r'Hi Team,', '', text, flags=re.IGNORECASE)
    text = re.sub(r'Hi Naresh/Prompt Team/ASK HR Team,', '', text, flags=re.IGNORECASE)
    text = re.sub(r'This is a follow-up to your previous request #\d+', '', text, flags=re.IGNORECASE)
    text = re.sub(r'Please find below Screenshot of course completion certificate ISDPAC FY \d+', '', text, flags=re.IGNORECASE)
    text = re.sub(r'This is a system generated mail. Please do not reply.', '', text, flags=re.IGNORECASE)

    # Remove specific phrases with Prism ID and training details
    text = re.sub(r'Prism is \d+', '', text)
    text = re.sub(r'Prism ID\s*:\s*\d+', '', text)
    text = re.sub(r'\[IS&DPA FY\d+\]', ' IS&DPA FY ', text)
    text = re.sub(r'\[DWP FY\d+\]', '', text)
    text = re.sub(r'\[POSH FY\d+\]', '', text)
    text = re.sub(r'\[TCOC FY\d+\]', '', text)
    text = re.sub(r'www\.[a-zA-Z0-9-]+\.[a-zA-Z]{2,}', '', text)




    # Remove specific phrases with phone numbers
    text = re.sub(r'\d{10}', '', text)

    # Remove email addresses
    text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', text)

    # Remove URLs
    text = re.sub(r'https?://\S+', '', text)

    # Remove extra spaces and newlines
    text = re.sub(r'\s+', ' ', text).strip()

    text = re.sub(r'&|:|<[^>]*>|fy\d{2}|\d{2}(?:jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)\d{2}|login\s*url\s*:|username\s*[:\s]+|password\s*[:\s]+|welcome@\d+', '', text, flags=re.IGNORECASE)

    # Remove specific substrings
    for substring in substrings_to_remove:
        text = text.replace(substring.lower(), '')  # Case-insensitive replacement

    return text

# Read the second CSV file into a pandas DataFrame
df_second = pd.read_csv('2nd part.csv')

# Apply the cleaning function to the "description" column and create a new column for cleaned description
df_second['cleaned_description'] = df_second['description'].apply(clean_description)

# Save the cleaned DataFrame to a new CSV file
df_second.to_csv('3rd part.csv', index=False)

print("Cleaning applied and saved to '3rd part.csv'")
```

# Cleaning the subject column

```python
import pandas as pd
import re
import numpy as np

def clean_subject(text):
    if isinstance(text, str):
        # Convert everything to lowercase
        cleaned_text = text.lower()

        # Remove 'Re:', dates, characters and numbers like [#165380], 'hi team', 'hi', '|', 'Subject:', square brackets, and curly brackets
        cleaned_text = re.sub(r're:\s*|\b\d{1,}\b\s*|\bhi\b\s*|hi\s*|\|', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r'\[.*?\]', '', cleaned_text)  # Remove square brackets and their contents
        cleaned_text = re.sub(r'\{.*?\}', '', cleaned_text)  # Remove curly brackets and their contents
        cleaned_text = re.sub(r'subject:\s*', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r'fwd:\s*', '', cleaned_text, flags=re.IGNORECASE)
        # cleaned_text = re.sub(r'*', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r'\*', '', cleaned_text)
        cleaned_text = re.sub(r'reg:\s*', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r'fw:\s*', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r'prism id:\s*', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r'recall:\s*', '', cleaned_text, flags=re.IGNORECASE)
        cleaned_text = re.sub(r're:\s*', '', cleaned_text, flags=re.IGNORECASE)


        return cleaned_text.strip()  # Strip any leading or trailing whitespace
    else:
        return np.nan  # Return NaN for non-string values


# Read the CSV file into a pandas DataFrame
df = pd.read_csv('3rd part.csv')

# Apply the cleaning function to the subject column and create a new column for cleaned subject
df['cleaned_subject'] = df['subject'].apply(clean_subject)

# Save the DataFrame with the cleaned subject column to a new CSV file
df.to_csv('4th part.csv', index=False)

print("Cleaning applied to subject column and saved to '4th part.csv'")
```

```python
from torch import cuda

model_id = 'meta-llama/Llama-2-7b-chat-hf'
device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'

print(device)

```

```python
from torch import bfloat16
import transformers

# set quantization configuration to load large model with less GPU memory
# this requires the `bitsandbytes` library

bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,  # 4-bit quantization
    bnb_4bit_quant_type='nf4',  # Normalized float 4
    bnb_4bit_use_double_quant=True,  # Second quantization after the first
    bnb_4bit_compute_dtype=bfloat16  # Computation type
)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Llama 2 Tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Llama 2 Model
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    quantization_config=bnb_config,
    device_map='auto',
)
model.eval()
```

```python
# Our text generator
generator = transformers.pipeline(
    model=model, tokenizer=tokenizer,
    task='text-generation',
    temperature=0.1,
    max_new_tokens=500,
    repetition_penalty=1.1
)
```

```python
prompt = "You are a Topic modelling & Text labelslling AI tool that "
res = generator(prompt)
print(res[0]["generated_text"])
```

```python
system_prompt = """
<s>[INST] <<SYS>>
Labeling topics and you perform topic modelling over data.
<</SYS>>
"""
```

```python
example_prompt = """
I have a topic that contains the following documents:

This is a follow-up to your previous request employment verification letter DeAndre McLeod. I was able to download the attached from prism. One of the other documents required for the Schengen visa is described as â€œdeclaration from your employer stating the period of absence in addition to the employment letter. I couldn't find this one in prism. I am including Christine huge and Kyoto Jain for quick hr approval so that p Srinivasan agreed to sign it and email it to me today. My visa appointment is tomorrow morning.

Output Topic - Background Verification
"""
```

```python
main_prompt = """
[INST]
I have a topic that contains the following documents:
[DOCUMENTS]

The topic is described by the following keywords: '[KEYWORDS]'.

Based on the information about the topic above, please create a short label of this topic.
[/INST]
"""
```

```python
prompt = system_prompt + example_prompt + main_prompt
```

```python
# import pandas as pd
from sentence_transformers import SentenceTransformer

# Pre-calculate embeddings
embedding_model = SentenceTransformer("BAAI/bge-small-en-v1.5")
embeddings = embedding_model.encode(abstracts, show_progress_bar=True)
```

```python
import pandas as pd

# Assuming 'abstracts' is a list of abstracts for which you've already calculated embeddings

# Create a DataFrame with abstracts and embeddings
df = pd.DataFrame({
    'abstract': abstracts,
    'embedding': embeddings.tolist()  # Convert embeddings from numpy array to list
})

# Save DataFrame to Parquet file
df.to_parquet('embeddings.parquet', index=False)
```

```python
from umap import UMAP
from hdbscan import HDBSCAN

# Initialize UMAP model
umap_model = UMAP(n_neighbors=15, n_components=5, min_dist=0.0, metric='cosine', random_state=1)
hdbscan_model = HDBSCAN(min_cluster_size=10, metric='euclidean', cluster_selection_method='leaf', prediction_data=True)

```

```python
# Pre-reduce embeddings for visualization purposes
reduced_embeddings = UMAP(n_neighbors=15, n_components=2, min_dist=0.0, metric='cosine', random_state=42).fit_transform(embeddings)
```

```python
import pandas as pd

# Assuming 'reduced_embeddings' is the reduced embeddings obtained from UMAP

# Create a DataFrame with reduced embeddings
df_reduced = pd.DataFrame(reduced_embeddings, columns=['UMAP_component_1', 'UMAP_component_2'])

# Save DataFrame to Parquet file
df_reduced.to_parquet('reduced_embeddings.parquet', index=False)
```

```python
from bertopic.representation import KeyBERTInspired, MaximalMarginalRelevance, TextGeneration

# KeyBERT
keybert = KeyBERTInspired()

# MMR
mmr = MaximalMarginalRelevance(diversity=0.3)

# Text generation with Llama 2
llama2 = TextGeneration(generator, prompt=prompt)

# All representation models
representation_model = {
    "KeyBERT": keybert,
    "Llama2": llama2,
    "MMR": mmr,
}
```

```python
import numpy as np
from bertopic import BERTopic
from hdbscan import HDBSCAN

embeddings = np.asarray(embeddings)

topic_model = BERTopic(
    # Sub-models
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model = HDBSCAN(min_samples=10),
    representation_model=representation_model,
    # Hyperparameters
    top_n_words=10,
    verbose=True
)

# Train model
topics, probs = topic_model.fit_transform(abstracts, embeddings)
```

```python
# Show topics
topic_model.get_topic_info()
```

```python
topic_info = topic_model.get_topic_info()
file_path = "topic_info.csv"
topic_info.to_csv(file_path, index=False)
```

```python
import pandas as pd

# Assuming topic_info is the output of topic_model.get_topic_info()

# Extracting topics and Llama2 values
topics = topic_info['Topic']
llama2_values = topic_info['Llama2']

# Creating a DataFrame
df = pd.DataFrame({'Topic': topics, 'Llama2': llama2_values})

# Displaying the DataFrame
print(df)
```
